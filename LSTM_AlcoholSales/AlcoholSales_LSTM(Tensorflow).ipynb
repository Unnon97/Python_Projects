# RNN model to predict Alcohol sales

Time series datasets provide us with information of a period of time and utilizing a Neural Network, specifically a Recurrent Neural Network helps us build a model that can predict the next data based on the trend. Here we work on building one such model that can help predict Alcohol sales by using previous n inputs

## About the dataset
The dataset is taken from Kaggle: Time series Forecasting, Alcohol Sales. Here we have two columns in the data: one shows the date from 1992 and other tells us the value of Alcohol sales monthly.



```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
import os
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
from tensorflow.keras.models import load_model
import pandas as pd
import torch.nn as nn
from statsmodels.tsa.seasonal import seasonal_decompose
from keras.preprocessing.sequence import TimeseriesGenerator
```

```python
df = pd.read_csv('data/Alcohol_Sales.csv', index_col = 'DATE', parse_dates = True)
df.index_freq = 'MS'
df.head()
```

```python
df.columns = ['Sales']
df.head()
```

```python
df.plot(figsize=(12,8))
```

```python
results = seasonal_decompose(df['Sales'])
```

```python
results.plot();
```

```python
datasize = len(df)
trainsize = int(datasize*0.8)
train = df.iloc[:trainsize]
test = df.iloc[trainsize:]
```

```python
#Scaling the data values to pass it into the RNN
scaler = MinMaxScaler()
scaler.fit(train)
scaled_train = scaler.transform(train)
scaled_test = scaler.transform(test)
```

```python
#Create TimeseriesGenerator object for training dataset and define the length of training sequence using n_input
n_input = 12
n_features = 1 # num of columns in data

train_generator = TimeseriesGenerator(scaled_train, scaled_train, length = n_input, batch_size=1)
```

```python
model = Sequential(
        LSTM(150, activation = 'relu', input_shape = (n_input, n_features)))
model.add(Dense(1))
model.compile(optimizer='adam', loss = 'mse')
```

```python
model.summary()
```

```python
model.fit(train_generator, epochs=100)
```

```python
model.history.history.keys()
```

```python
#Plotting loss versus the range of values
myloss = model.history.history['loss']
plt.plot(range(len(myloss)), myloss)
```

```python
#For predicting the first data in test set, we require the previous 12 data points from the trained dataset
first_batch = scaled_train[-12:]
print(first_batch)
print(first_batch.shape)
```

```python
# Reshaping the first_batch as input passed to TimeseriesGenerator must be of shape (batch_size, n_input, n_features)
first_batch = first_batch.reshape((1,n_input,n_features))
model.predict(first_batch)
```

```python

test_predictions = []
first_evalbatch = scaled_train[-n_input:]
current_batch = first_evalbatch.reshape((1,n_input,n_features))
```

```python
for i in range(len(test)):
    current_pred = model.predict(current_batch)[0]
    #store prediction
    test_predictions.append(current_pred)
    current_batch = np.append(current_batch[:,1:,:], [[current_pred]], axis = 1)
```

```python
true_preditions = scaler.inverse_transform(test_predictions)
print(true_preditions)
#Add Prediction column name to the data
test['Predictions'] = true_preditions
```

```python
#Plotting original values with prediction to compare
test.plot(figsize=(12,8))
```

```python
#Saving model for future reference
model.save('timeseriesmodel.h5')
timeseries_rnn_model = load_model('timeseriesmodel.h5')
timeseries_rnn_model.summary()
```
